```python
df['dep_hour'] = df['Dep_Time'].str.split(":").str[0]
df['dep_mins'] = df['Dep_Time'].str.split(":").str[1]
df = df.drop("Dep_Time", axis=1)
```

```python
df['dep_hour']=df['dep_hour'].astype(int)
df['dep_mins']=df['dep_mins'].astype(int)
```
